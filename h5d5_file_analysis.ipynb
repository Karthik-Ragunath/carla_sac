```python

def print_hdf5_item(name, obj):
    if isinstance(obj, h5py.Dataset):
        print("Dataset:", name)
    elif isinstance(obj, h5py.Group):
        print("Group:", name)
```

```python
# Open the HDF5 file
file_path = '0002.h5'
with h5py.File(file_path, 'r') as file:
    # Traverse the file structure recursively
    file.visititems(print_hdf5_item)
```
